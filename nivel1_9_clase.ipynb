**Matriz de correlaciones**

```python
import numpy as np
import math as math
import scipy.stats as stats

df_corr = df_cont.corr(method="pearson")
#df_corr = df_cont.corr()
df_corr
```

```python
n = df_cont.shape[0]
p = df_cont.shape[1]
```

```python
chi2 = -(n-1-(2*p+5)/6)*math.log(np.linalg.det(df_corr))
```

```python
chi2
```

```python
ddl = p*(p-1)/2
```

```python
p_valor = stats.chi2.pdf(chi2,ddl)
p_valor
```

**Las variables sí están correlacionadas**

```python
import numpy as np
import matplotlib.pyplot as plt

cov_mat = np.cov(X_train_std.T) #se calcula la matriz de covarianza del conjunto de datos estandarizado
```

**Varianza explicada**

```python
from sklearn.decomposition import PCA
```

```python
sum(pca.explained_variance_ratio_[0:6])
#varianza explicada por los 6 componentes principales que indica el criterio de Kaiser
```
